```python
# 1. ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¡œë“œ
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from google.colab import files

# 2. ì—‘ì…€ íŒŒì¼ ì—…ë¡œë“œ
uploaded = files.upload()
df_all = pd.read_excel("ì²œì›ì˜ì•„ì¹¨ë°¥ë°ì´í„°ì…‹.xlsx", sheet_name=None)

# 3. ì‹œíŠ¸ í†µí•© ë° ì „ì²˜ë¦¬
data = []
for month, df in df_all.items():
    df = df.dropna(how='all').reset_index(drop=True)
    for _, row in df.iterrows():
        date = row.get('ë‚ ì§œ')
        if pd.isna(date): continue
        for col in df.columns[1:6]:
            menu = row.get(col)
            if pd.notna(menu):
                data.append({'ë‚ ì§œ': date, 'ì›”': month, 'ë©”ë‰´': menu})

df = pd.DataFrame(data)
df['ë‚ ì§œ'] = pd.to_datetime(df['ë‚ ì§œ'].astype(str), format='%y.%m.%d', errors='coerce')
df = df.dropna(subset=['ë‚ ì§œ'])
df['ìš”ì¼'] = df['ë‚ ì§œ'].dt.day_name()

# 4. ì„ì˜ ìƒì„±: ì‹ì‚¬ ì¸ì›, ì”ë°˜ëŸ‰, ë©”ë‰´ ì„ í˜¸ë„
np.random.seed(42)
df['ì‹ì‚¬ ì¸ì›'] = np.random.randint(150, 300, size=len(df))
df['ì”ë°˜ëŸ‰(kg)'] = np.round(np.random.uniform(3.0, 25.0, size=len(df)), 1)
df['ë©”ë‰´ ì„ í˜¸ë„(%)'] = np.round(np.random.uniform(60.0, 95.0, size=len(df)), 1)

# 5. ì¸ì½”ë”© ë° X, y ìƒì„±
X_raw = df[['ìš”ì¼', 'ë©”ë‰´', 'ì‹ì‚¬ ì¸ì›', 'ë©”ë‰´ ì„ í˜¸ë„(%)']]
y = df['ì”ë°˜ëŸ‰(kg)']

encoder = OneHotEncoder(sparse_output=False)
X_cat = encoder.fit_transform(X_raw[['ìš”ì¼', 'ë©”ë‰´']])
X_num = X_raw[['ì‹ì‚¬ ì¸ì›', 'ë©”ë‰´ ì„ í˜¸ë„(%)']].values
X = np.hstack([X_cat, X_num])

# 6. train/test split (index ë³´ì¡´ ë°©ì‹)
indices = np.arange(len(X))
train_idx, test_idx = train_test_split(indices, test_size=0.2, random_state=42)
X_train, X_test = X[train_idx], X[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# 7. ê²°ê³¼ ì €ì¥ìš© ë¦¬ìŠ¤íŠ¸
results = []

# âœ… RandomForest
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
results.append(['RandomForest',
                mean_absolute_error(y_test, rf_pred),
                np.sqrt(mean_squared_error(y_test, rf_pred)),
                r2_score(y_test, rf_pred)])

# âœ… XGBoost
xgb = XGBRegressor(random_state=42)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
results.append(['XGBoost',
                mean_absolute_error(y_test, xgb_pred),
                np.sqrt(mean_squared_error(y_test, xgb_pred)),
                r2_score(y_test, xgb_pred)])

# âœ… LSTM
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_lstm = X_scaled.reshape(-1, 1, X_scaled.shape[1])
X_train_lstm, X_test_lstm = X_lstm[train_idx], X_lstm[test_idx]

lstm_model = Sequential()
lstm_model.add(Dense(64, activation='relu', input_shape=(1, X.shape[1])))
lstm_model.add(Dense(32, activation='relu'))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

lstm_model.fit(X_train_lstm, y_train, epochs=30, batch_size=16, verbose=0)
lstm_pred = lstm_model.predict(X_test_lstm).flatten()

results.append(['LSTM',
                mean_absolute_error(y_test, lstm_pred),
                np.sqrt(mean_squared_error(y_test, lstm_pred)),
                r2_score(y_test, lstm_pred)])

# ğŸ“Š ìµœì¢… ì„±ëŠ¥ ë¹„êµí‘œ ì¶œë ¥
result_df = pd.DataFrame(results, columns=['ëª¨ë¸', 'MAE', 'RMSE', 'RÂ²'])
print("ğŸ“Š ì„±ëŠ¥ ë¹„êµ ê²°ê³¼:")
print(result_df)
```
